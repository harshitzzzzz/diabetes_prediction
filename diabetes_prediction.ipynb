```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

**data collection and analysis**


```python
dataset=pd.read_csv("diabetes.csv")
```

```python
dataset
```

```python
x=dataset.drop(columns="Outcome",axis=1)
y=dataset["Outcome"]
```

```python
x
```

```python
y
```

**data standardization**

```python
scaler=StandardScaler()
```

```python
scaler.fit(x)
```

```python
standardized=scaler.transform(x)
```

```python
standardized
```

```python
x=standardized
y=dataset["Outcome"]
```

**split and train**

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
x.shape
```

```python
xtrain.shape,xtest.shape
```

**training**

```python
classifier=svm.SVC(kernel='linear')
```

```python
classifier.fit(xtrain,ytrain)
```

**evaluation**

```python
xtrainPrediction=classifier.predict(xtrain)
training_data_accuracy=accuracy_score(xtrainPrediction,ytrain)
```

```python
training_data_accuracy
```

```python
xtestPrediction=classifier.predict(xtest)
test_data_accuracy=accuracy_score(xtestPrediction,ytest)
```

```python
test_data_accuracy
```

```python
input_data=(1,189,60,23,846,30.1,0.398,59)
array=np.array(input_data)
input_reshape=array.reshape(1,-1)
std_data=scaler.transform(input_reshape)
```

```python
prediction=classifier.predict(std_data)
if(prediction[0]==0):
    print(" no diabetes")
else:
    print("has diabetes")
```
